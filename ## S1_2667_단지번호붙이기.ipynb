```python
n = int(input())
mp = [list(input()) for _ in range(n)]
cnt_list = []
for i in range(n):
    for j in range(n):
        if mp[i][j]=='1':
            cnt=0
            work = [(i,j)]
            while work:
                x,y = work.pop(0)
                if mp[x][y]=='1':
                    cnt+=1
                    mp[x][y]=False
                    for dx, dy in [(1,0),(0,1),(-1,0),(0,-1)]:
                        dx += x
                        dy += y
                        if 0<=dx<n and 0<=dy<n and mp[dx][dy]=='1':
                            work.append((dx,dy))
            cnt_list.append(cnt)
print(len(cnt_list))
print(sorted(cnt_list, reverse=True), sep='\n')
```

고수풀이

```python
h = [0]
l = range(n:=int(input()))
m = [[*input()] for _ in l]

def f(x,y):
    if 0<=x<n and 0<=y<n:
        if m[x][y]=='0': return
        m[x][y] = '0'
        f(x+1,y)
        f(x,y+1)
        f(x-1,y)
        f(x,y-1)
        h[h[0]] += 1

for r in l:
    for c in l:
        if m[r][c]=='1':
            h[0] += 1
            h = h+[0]
            f(r,c)

print(h[0])
print(*sorted(h[1:]))
```
